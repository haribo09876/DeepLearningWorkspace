# 5. 시계열 데이터와 딥러닝 네트워크

- 시계열 데이터는 음성, 동영상 등과 같이 각종 센서에서 일정 시간 간격으로, 지속적으로 측정되는 데이터
- 시계열 데이터의 측정은 자이로센서, 압력 센서, 마이크, 카메라 등의 각종 센서를 사용할 수 있음
- 시계열 데이터의 이름은 데이터가 시간의 흐름에 따라 변화한다는 의미
- 티쳐블 머신을 음성/사운드 데이터의 인식에 사용할 수 있음
- 시계열 데이터가 일반적으로 가지는 특성은 1) 빅데이터, 2) 반복되는 패터, 3) 결측치/이상치, 4) 잡음
- 데이터 특성은 데이터의 종류에 따라 다르며, 이 중 일부 특성만 가지는 경우가 많음
- 시계열 데이터의 학습에 딥러닝을 사용하기 위해서는 데이터 내에 존재하는 결측치, 이상치, 잡음을 제거하거나 보완해야 함
- 시계열 데이터의 전처리 과정에서 시계열 데이터의 반복적인 특성을 활용할 수 있음
- 신호를 분할하여 사용하면 데이터의 크기가 감소하여 학습이 용이해짐
- LSTM층은 시계열데이터가 변화하는 양상을 모델링할 수 있는 네트워크 층임
- 시계열 데이터의 학습을 위해 1차원 합성곱층을 함께 사용할 수 있음
