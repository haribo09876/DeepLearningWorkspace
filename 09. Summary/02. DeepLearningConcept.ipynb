# 2. 딥러닝의 개념

- 딥러닝은 인공지능의 오래된 학습 이론이며, 병렬처리 기술, GPU 가속 기술 등의 성능 향상과 함께 인공지능기술의 주류가 됨
- '딥러닝 모델'은 입력이 주어졌을 때, 그 입력에 대한 해석을 출력하는 일종의 '블랙박스' 임
- 딥러닝은 기계학습 이론의 일종임
- 컴퓨터에 있어 학습은 주어진 데이터를 단순히 기억하는 것이 아니라, 해당 데이터를 일반화하는 모델을 만드는 것
- 딥러닝 모델은 서로 간에 데이터를 주고받을 수 있는 여러 개의 네트워크층(layer)으로 이루어져 있음
- 딥러닝 모델의 네트워크층들은 크게 입력층, 중간층, 출력층으로 나누어 살펴볼 수 있으며, 각각의 층들은 데이터를 입력 받아 변형하고, 변형된 데이터를 다음층으로 전달하는 구조
- 복잡하고 많은 수의 데이터를 인식하기 위해, 네트워크층들은 서로 복잡하게 연결되도록 구성될 수 있음
- 네트워크의 입력과 출력은 모두 숫자를 사용하여 이루어짐
- 범주형 데이터는 원-핫 인코딩 방법을 통해 숫자로 변환되어 사용
- 원-핫-인코딩은 범주의 개수만큼 항목을 만들고, 각각의 항목이 각 범주를 나타내도록 하는 데이터의 표현 방법
- 숫자로 표현된 범주형 데이터도 있음
- 수치형 데이터는 0과 1 사이의 범위로 정규화하여 사용하는 것이 일반적
- 수치형 데이터를 0과 1 사이의 값으로 정규화하려면 해당 데이터의 최댓값, 최솟값을 알아야 함
- 데이터의 최대/최소 추정치를 사용하여 수치형 데이터를 정규화할 수도 있음
