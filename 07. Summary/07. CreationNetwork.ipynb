# 7. 생성 네트워크

- 생성 네트워크는 주어진 파라미터로부터 해당 파라미터와 연관된 데이터를 자동으로 만드는 네트워크임
- 생성 네트워크는 특정한 형상을 가지는 소리, 이미지, 비디오 등을 생성하는 데에 사용됨
- 생성 네트워크를 구성하기 위해 Reshape층, Upsampling층 등이 사용됨
- Reshape층은 Flatten층의 역연산으로, Upsampling층은 풀링층의 역연산으로 생각할 수 있음
- GAN(적대적 생성 네트워크)는 두 네트워크 간의 상호 경쟁을 통해 네트워크의 성능을 향상시키는 개념을 구현한 것
- GAN은 판별자와 생성자의 두 개 네트워크로 구성됨
- 판별자는 합성곱 신경망, 순환 신경망 등의 일반적인 '분류 네트워크'를 사용하며, 생성자는 '생성 네트워크'를 사용
- 안정적인 GAN의 구현을 위해 배치 정규화 층이 활용될 수 있음
- 배치 정규화 층은 합성곱 신경망에서 오버피팅을 감소시키기 위해 사용되기도 함
