# 8. 강화학습

- 강화 학습은 특정 입력에 대해 지정된 정답을 알 수 없는 경우를 위해 개발된 학습 방법으로, 주어진 상황에서 최적의 행동을 도출하는 것을 그 목표로 함
- 강화 학습은 전략 게임, 바둑, 체스 등의 보드게임의 인공지능 구현에 널리 사용되고 있으며, 로봇/드론의 행동 제어, 전투기의 조종 등과 같은 실제 환경에서의 제어 문제에도 활용되어 뛰어나느 성능을 보이고 있음
- 게임 캐릭터/자동차 등의 제어 문제를 일반적인 지도 학습으로 풀기 어려운 이유는 특정 상황에서의 올바른 움직임을 정의하는 것 자체가 매우 어렵다는 것
- 강화 학습은 환경(State)과 에이전트(Agent)의 상호작용을 기반으로 함
- 에이전트의 행동(지시)에 따라 변화하는 환경을 디자인하고, 에이전트가 최적의 결과를 내도록 학습
- 환경은 그 종류에 따라 에이전트와 동기화되어 운영될 수도, 비동기식으로 운영될 수도 있음
- 강화 학습에서 에이전트는 시뮬레이션을 수행 후 보상을 획득
- 할인율을 획득하는 보상의 지연에 따라 손실되는 보상의 정도를 나타냄
