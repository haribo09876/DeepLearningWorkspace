# 4. 이미지를 인식하는 네트워크, 합성곱 신경망

- 티쳐블 머신(Teachable Machine)은 이미지 인식 등을 위한 웹상에서 제공되는 무료 인공지능 도구임
- 마우스 클릭 몇 번만으로 구성할 수 있는 기본 모델들을 제공하고 있으며, 몇 장의 사진을 추가하여 손쉽게 인공지능을 경험하고 사용할 수 있음
- 합성곱(Convolution)은 두 개의 행렬에 대한 연산 방법 중 하나임
- 합성곱은 두 행렬의 값을 원소 단위로 곱하여 더하는 것으로, 행렬 원소 각각의 값들과 그 주변 값들과의 관계를 고려하여 계산을 하게 된다는 특징이 있음
- 합성곱 연산을 통해 에지 정보와 같은 이미지의 특징을 추출할 수 있음
- 합성곱층은 딥러닝 네트워크에서 합성곱 연산이 이루어지는 층
- 합성곱층과 완전 연결층을 사용하여 신경망을 구성할 수 있고, 이를 합성곱 신경망이라고 부름
- 합성곱 신경망에서 합성곱층에 대한 학습은 그림의 특징을 추출하는 필터를 자동으로 찾아내는 과정이라고 볼 수 있음
- 합성곱 신경망에서 풀링층은 특징의 개수, 학습해야 할 데이터의 크기를 줄여 주는 역할을 함
- 풀링층의 종류에는 Max, Average, Min 등이 있음
- 풀링층을 구성할 때 Pool size와 Stride를 설정해 주어야 함
- Flatten층은 합성곱층과 완전 연결층을 연결하기 위해 필요하고, 두 층 사이에 두고 사용
